```python
# Some terms that share similarities and may be confusing at first:

# Business analytics
# Data analytics
# Data Science
# Business Intelligence
# Machine learning


# Analysis VS Analytics (Analysis = Analytics?):

# Analysis: This is when you separate the datasets into easily todigest chunks
# amd study them individually and examine how they relate to other parts. One
# thing to remember is that we perform analysis on things that have already
# happened (in the past) such as using an analysis to explain how a story ended
# the way it did or to explain a decline in sales.

# Analytics: This generally refers to the future. Instead of explaining past 
# events, it explores potential future ones. This is the application of logical
# and computational reasoning to the component parts obtained in an analysis.
# We look for patterns and explore what we can do with them in the future.

# Analytics branches off into two areas:

# Qualitative analytics: Here we use our intuition and experience in conjuction
# with the analysis in order to plan our next business move

# Quantitative analytics: This is applying formulas and algorithms to numbers
# we have gathered from our analysis.


```

```python
# Examples of analytics:

# Qualitative:

# As the owner of a clothing store, you have performed detailed analysis
# of what your customers like and you feel sure about which fashion trends to
# to follow. So, upon studying a women's clothes magazine you decide to introd-
# uce a certan clothing line to your store.

#Quantitative:

# Suppose that you want to introduce a new clothing line to your store, but
# you don't know which month would be the best to do that, so you look at
# past sales data by month to decide in which month would be best to introduce
# the new clothing line.
```
